## Data Preprocessing
### We can convert Latitude and Longitude into addresses which will be much easier for us to understand

```python
pip install geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent = 'geoapiExercises')
```

```python
geolocator.reverse("37.88"+" , "+"-122.23" ).raw['address']
```

```python
def location(cord):
    latitude = str(cord[0])
    longitude = str(cord[1])
    
    location = geolocator.reverse( latitude + "," + longitude).raw['address']
    
    if location.get('road') is None:
        location['road'] = None
    
    if location.get('county') is None:
        location['county'] = None
    
    loc_update['county'].append(location['county'])
    loc_update['road'].append(location['road'])
```

```python
loc_update = {"county": [],
             "road": []}

for i,cord in enumerate(df.iloc[:,6:-1].values):
    location(cord)
    
    pickle.dump(loc_update, open('loc_update.pickle','wb'))
    
    if i % 100 == 0:
        print(i)
```

```python
loc_update = pickle.load(open("loc_update.pickle", "rb"))
```

```python
loc_update
```

```python
loc = pd.DataFrame(loc_update)
```

```python
loc.info()
```

## Add the new features to the dataframe

```python
for i in loc_update.keys():
    df[i] = loc_update[i]
    
df = df.sample(axis = 0, frac = 1)
df.head(10)
```

### Drop latitude, longitude column

```python
df = df.drop(labels = ["Latitude", "Longitude"], axis = 1)
df.head()
```

### Classification algorithm to find missing Categorical value

```python
missing_idx = []
for i in range(df.shape[0]):
    if df['road'][i] is None:
        missing_idx.append(i)
        
missing_road_X_train = np.array([[df['MedInc'][i], df['AveBedrms'][i], df['AveRooms'][i]] for i in range(df.shape[0]) if i not in missing_idx])

missing_road_Y_train = np.array([df['road'][i] for i in range(df.shape[0]) if i not in missing_idx])

missing_road_X_test = np.array([[df['MedInc'][i], df['AveBedrms'][i], df['AveRooms'][i]] for i in range(df.shape[0]) if i in missing_idx])
```

```python
missing_road_X_train
```

```python
missing_road_Y_train
```

```python
missing_road_X_test.shape
```

### We Use Stochastic Gradient Classifier

```python
from sklearn.linear_model import SGDClassifier
```

#### Model Training

```python
model_1 = SGDClassifier()
model_1.fit(missing_road_X_train, missing_road_Y_train)
```

```python
missing_road_y_pred = model_1.predict(missing_road_X_test)
```

```python
np.unique(missing_road_y_pred)
```

### Add the model back to the dataframe

```python
for n, i in enumerate(missing_idx):
    df['road'][i] = missing_road_y_pred[n]
```

```python
df.info()
```

### Label Encode on the Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['road'] = le.fit_transform(df['road'])
```

```python
df.head()
```

### Classification algorithm to find missing Categorical value(county) same as we did for road

```python
missing_idx = []
for i in range(df.shape[0]):
    if df['county'][i] is None:
        missing_idx.append(i)
        
missing_county_X_train = np.array([[df['MedInc'][i], df['AveBedrms'][i], df['AveRooms'][i]] for i in range(df.shape[0]) if i not in missing_idx])

missing_county_Y_train = np.array([df['county'][i] for i in range(df.shape[0]) if i not in missing_idx])

missing_county_X_test = np.array([[df['MedInc'][i], df['AveBedrms'][i], df['AveRooms'][i]] for i in range(df.shape[0]) if i in missing_idx])
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
model_1 = SGDClassifier()
model_1.fit(missing_county_X_train, missing_county_Y_train)
```

```python
missing_county_y_pred = model_1.predict(missing_county_X_test)
```

```python
np.unique(missing_county_y_pred)
```

```python
for n, i in enumerate(missing_idx):
    df['county'][i] = missing_county_y_pred[n]
```

```python
df.info()
```

```python
le = LabelEncoder()
df['county'] = le.fit_transform(df['county'])
```

```python
df.head()
```

## Train-Test Split

```python
y = df.iloc[:,-3].values
df.drop(labels= ['Target'], axis= 1, inplace = True)
x = df.iloc[:,:].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

## Model Selection

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

## Checking the accuracy of the model

```python
from sklearn.metrics import r2_score

accuracy = r2_score(y_test, y_pred)
```

```python
ACCURACY = accuracy * 100
print("Accuracy of the model is: ",ACCURACY)
```
